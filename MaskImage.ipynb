```python
import cv2
import os
import numpy as np

# Input and output paths
video_path = "Masking_Image/test9_cs412-21s.mp4"
output_dir = "masking_out_directory"

# Ensure video file exists
if not os.path.isfile(video_path):
    print(f"Error: {video_path} not found.")
    exit()

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the video file
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Validate FPS
if fps <= 0:
    print(f"Error: Could not retrieve FPS from {video_path}")
    cap.release()
    exit()

frame_interval = int(fps)  # Extract a frame every second
index = 0
count = 0
success, frame = cap.read()

while success:
    if count % frame_interval == 0:
        img_file = os.path.join(output_dir, f"scene_{index}.png")
        
        # Apply masking to the extracted frame
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        lower_hsv = np.array([0, 0, 120])
        upper_hsv = np.array([180, 38, 255])
        mask = cv2.inRange(hsv, lower_hsv, upper_hsv)
        result = cv2.bitwise_and(frame, frame, mask=mask)
        _, g, _ = cv2.split(result)
        filtered = g.copy()

        # Apply binary threshold
        _, binary_mask = cv2.threshold(filtered, 10, 255, cv2.THRESH_BINARY)

        # Apply background removal
        bg_removed = frame.copy()
        bg_removed[binary_mask != 0] = [255, 255, 255]  # Set background to white
        
        # Save the processed image
        cv2.imwrite(img_file, bg_removed)
        print(f"Saved masked frame: {img_file}")
        
        index += 1
    
    success, frame = cap.read()
    count += 1

cap.release()
print("Processing complete.")
```

```python
# Contour Masking
import cv2
import os
import numpy as np

# Input and output paths
video_path = "data/videos/test0_toy_lecture.mp4"
output_dir = "pixelsubtract_directory2"

# Ensure video file exists
if not os.path.isfile(video_path):
    print(f"Error: {video_path} not found.")
    exit()

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the video file
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Validate FPS
if fps <= 0:
    print(f"Error: Could not retrieve FPS from {video_path}")
    cap.release()
    exit()

frame_interval = int(fps)  # Extract a frame every second
index = 0
count = 0
success, frame = cap.read()

previous_canvas = None  # Store the previous frame's contours

while success:
    if count % frame_interval == 0:
        img_file = os.path.join(output_dir, f"scene_{index}.jpg")

        # Convert frame to grayscale
        image_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Convert grayscale to binary (black and white)
        _, image_edges = cv2.threshold(image_gray, 100, 255, cv2.THRESH_BINARY)

        # Create white canvas, mask, and background
        canvas = np.full(frame.shape, 255, np.uint8)
        mask = np.full(frame.shape, 255, np.uint8)
        new_background = np.full(frame.shape, 255, np.uint8)

        # Find contours
        contours_draw, _ = cv2.findContours(image_edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        contours_mask, _ = cv2.findContours(image_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

        # Draw all contours on the canvas
        for contour in range(len(contours_draw)):
            cv2.drawContours(canvas, contours_draw, contour, (0, 0, 0), 3)

        # Create mask and background
        for contour in range(len(contours_mask)):
            if contour != 1:
                cv2.fillConvexPoly(mask, contours_mask[contour], (0, 0, 0))
                cv2.fillConvexPoly(new_background, contours_mask[contour], (0, 255, 0))  # Green background

        # Apply the mask to remove the background
        output_frame = cv2.bitwise_and(frame, new_background)

        # Save the processed images
        cv2.imwrite(img_file, output_frame)
        contours_file = os.path.join(output_dir, f"contours_{index}.png")
        mask_file = os.path.join(output_dir, f"mask_{index}.png")
        background_file = os.path.join(output_dir, f"background_{index}.png")

        cv2.imwrite(contours_file, canvas)
        cv2.imwrite(mask_file, mask)
        cv2.imwrite(background_file, new_background)

        print(f"Saved masked frame: {img_file}")

        # Compute difference with the previous contour frame and save
        if previous_canvas is not None:
            diff = cv2.absdiff(previous_canvas, canvas)  # Subtract adjacent frames
            diff_file = os.path.join(output_dir, f"contour_diff_{index-1}_{index}.png")
            cv2.imwrite(diff_file, diff)
            print(f"Saved difference frame: {diff_file}")

        # Update previous canvas
        previous_canvas = canvas.copy()

        index += 1

    success, frame = cap.read()
    count += 1

cap.release()
print("Processing complete.")
```

```python
# Scrolling 1st attempt
import cv2
import os
import numpy as np

# Input and output paths
video_path = "data/videos/scrolling.mov"
output_dir = "scrolling_directory4"

# Ensure video file exists
if not os.path.isfile(video_path):
    print(f"Error: {video_path} not found.")
    exit()

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the video file
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Validate FPS
if fps <= 0:
    print(f"Error: Could not retrieve FPS from {video_path}")
    cap.release()
    exit()

frame_interval = int(fps)  # Extract a frame every second
index = 0
count = 0
success, prev_frame = cap.read()

scrolling = False  # Track scrolling state
motion_threshold = None  # Will be calculated dynamically

# Compute initial motion threshold using the first 30 frames
motion_values = []
for _ in range(30):
    success, frame = cap.read()
    if not success:
        break
    gray_prev = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    gray_curr = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frame_diff = cv2.absdiff(gray_prev, gray_curr)
    motion_values.append(np.mean(frame_diff))  # Store mean difference
    prev_frame = frame.copy()

# Set the motion threshold dynamically
motion_threshold = np.mean(motion_values) * 2.5  # Double the average motion
print(f"Dynamic motion threshold set to: {motion_threshold}")

# Reset video reading
cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
success, prev_frame = cap.read()

while success:
    success, frame = cap.read()
    if not success:
        break  # Stop if the video ends

    # Convert frames to grayscale for motion detection
    gray_prev = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    gray_curr = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Compute absolute difference between consecutive frames
    frame_diff = cv2.absdiff(gray_prev, gray_curr)
    diff_value = np.mean(frame_diff)  # Mean difference

    print(f"Frame {index} - Motion Level: {diff_value}")

    if diff_value > motion_threshold:
        if not scrolling:
            print(f"Scrolling Detected at frame {index} - Applying Mask")
        scrolling = True
    else:
        if scrolling:
            print(f"Scrolling Stopped at frame {index} - Removing Mask")
        scrolling = False

    # Apply mask only if scrolling is detected
    if scrolling:
        mask = np.full(frame.shape, (0, 255, 0), dtype=np.uint8)  # Green overlay
        frame = cv2.addWeighted(frame, 0.7, mask, 0.3, 0)  # Blend mask with video

    # Save the processed frame
    img_file = os.path.join(output_dir, f"masked_{index}.jpg")
    cv2.imwrite(img_file, frame)

    prev_frame = frame.copy()
    index += 1

cap.release()
print("Processing complete.")
```

```python
# Scrolling 4th Attempt

import cv2
import os
import numpy as np

# Input and output paths
video_path = "data/videos/scrolling.mov"
output_dir = "scrolling_directory7"

# Ensure video file exists
if not os.path.isfile(video_path):
    print(f"Error: {video_path} not found.")
    exit()

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the video file
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Validate FPS
if fps <= 0:
    print(f"Error: Could not retrieve FPS from {video_path}")
    cap.release()
    exit()

index = 0
success, prev_frame = cap.read()

scrolling = False  # Track scrolling state
motion_history = []  # Stores last few motion values for smoothing
rolling_window = 10  # Shorter window to make the threshold more adaptive

# Function to compute frame motion
def compute_motion(prev_frame, curr_frame):
    """Computes a motion value based on edge and intensity differences."""
    gray_prev = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    gray_curr = cv2.cvtColor(curr_frame, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    gray_prev = cv2.GaussianBlur(gray_prev, (5, 5), 0)
    gray_curr = cv2.GaussianBlur(gray_curr, (5, 5), 0)

    # Compute Canny edges
    edges_prev = cv2.Canny(gray_prev, 50, 150)
    edges_curr = cv2.Canny(gray_curr, 50, 150)

    # Compute absolute difference between intensity values
    intensity_diff = cv2.absdiff(gray_prev, gray_curr)
    edge_diff = cv2.absdiff(edges_prev, edges_curr)

    # Compute weighted motion score
    motion_value = (np.sum(intensity_diff) * 0.6 + np.sum(edge_diff) * 0.4) / 1000  # Normalize
    return motion_value

# Compute initial motion threshold using the first few frames
for _ in range(rolling_window):
    success, frame = cap.read()
    if not success:
        break
    motion_value = compute_motion(prev_frame, frame)
    motion_history.append(motion_value)
    prev_frame = frame.copy()

# Compute adaptive threshold using percentile-based filtering
motion_threshold = np.percentile(motion_history, 60) * 1.1  # Use a lower threshold
print(f"Adaptive motion threshold set to: {motion_threshold}")

# Reset video reading
cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
success, prev_frame = cap.read()

while success:
    success, frame = cap.read()
    if not success:
        break  # Stop if the video ends

    # Compute frame motion
    diff_value = compute_motion(prev_frame, frame)

    # Maintain rolling window for threshold adjustment
    motion_history.append(diff_value)
    if len(motion_history) > rolling_window:
        motion_history.pop(0)  # Remove old values

    # Dynamically update threshold using 60th percentile
    dynamic_threshold = np.percentile(motion_history, 60) * 1.1

    print(f"Frame {index} - Motion Level: {diff_value}, Dynamic Threshold: {dynamic_threshold}")

    # Detect scrolling based on updated threshold
    if diff_value > dynamic_threshold:
        if not scrolling:
            print(f"Scrolling Detected at frame {index} - Applying Mask")
        scrolling = True
    else:
        if scrolling:
            print(f"Scrolling Stopped at frame {index} - Removing Mask")
        scrolling = False

    # Apply mask only if scrolling is detected
    if scrolling:
        mask = np.full(frame.shape, (0, 255, 0), dtype=np.uint8)  # Green overlay
        frame = cv2.addWeighted(frame, 0.7, mask, 0.3, 0)  # Blend mask with video

    # Save the processed frame
    img_file = os.path.join(output_dir, f"masked_{index}.jpg")
    cv2.imwrite(img_file, frame)

    prev_frame = frame.copy()
    index += 1

cap.release()
print("Processing complete.")
```
